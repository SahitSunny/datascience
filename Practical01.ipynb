```python
df['Age'] = df['Age'].values.astype(dtype = int)
```

<h2> Normalization </h2>

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
fare_scaled = scaler.fit_transform(df[['Fare']])
fare_scaled
```

```python
fare_scaled = pd.DataFrame(fare_scaled, columns=scaler.get_feature_names_out(['Fare']))
```

```python
fare_scaled
```

<h2> Encoding </h2>

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop = 'first')
encoded = ohe.fit_transform(df[['Embarked']])
encoded
```

```python
encoded = pd.DataFrame(encoded.toarray(), columns=ohe.get_feature_names_out(['Embarked']))
```
