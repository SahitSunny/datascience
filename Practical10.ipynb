```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('IRIS.csv')
df
```

```python
plt.figure(figsize=(10, 8))
df.hist()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data= df)
```

```python
numeric_columns = df.iloc[: , :-1]
```

```python
numeric_columns
```

```python
Q1 = numeric_columns.quantile(0.25)
Q3 = numeric_columns.quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = ((numeric_columns < lower_limit) | (upper_limit < numeric_columns)).any(axis= 1)

print(f"Indices : {df[outliers]}")
print(f"Count : {outliers.sum()}")
```
